```python
from transformers import pipeline

# 1. Класифікація тональності (sentiment analysis)
sentiment_analyzer = pipeline("sentiment-analysis")
text1 = "Я дуже задоволений сервісом, все було швидко і якісно."
result1 = sentiment_analyzer(text1)[0]

# 2. Витяг сутностей (NER)
ner_extractor = pipeline("ner", grouped_entities=True)
text2 = "У 2020 році Google придбав компанію Fitbit."
result2 = ner_extractor(text2)

# 3. Переклад (Ukr -> En)
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-uk-en")
text3 = "Мені подобається програмувати."
result3 = translator(text3)[0]['translation_text']

# 4. Генерація тексту
generator = pipeline("text-generation", model="gpt2")
text4 = "Напиши короткий опис для сайту оренди авто:"
result4 = generator(text4, max_length=40, num_return_sequences=1)[0]['generated_text']

# 5. Питання-відповідь (QA)
qa_pipeline = pipeline("question-answering")
context = "Україна стала незалежною у 1991 році. Першим президентом був Леонід Кравчук."
question = "Хто був першим президентом незалежної України?"
result5 = qa_pipeline(question=question, context=context)

# Виведення результатів
print("1. Тональність:", result1)
print("2. Сутності:", result2)
print("3. Переклад:", result3)
print("4. Генерація:", result4)
print("5. QA:", result5)
```

```python
# ====== Гібридна рекомендаційна система для MovieLens в Colab ======

import pandas as pd
import zipfile
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# -------------------------------
# 1. Розпаковка архіву
# -------------------------------
with zipfile.ZipFile("ml-latest-small.zip", 'r') as zip_ref:
    zip_ref.extractall("ml-small")

ratings = pd.read_csv("ml-small/ml-latest-small/ratings.csv")
movies = pd.read_csv("ml-small/ml-latest-small/movies.csv")

# -------------------------------
# 2. Приводимо movieId до int
# -------------------------------
ratings['movieId'] = ratings['movieId'].astype(int)
movies['movieId'] = movies['movieId'].astype(int)

# -------------------------------
# 3. Content-Based підготовка
# -------------------------------
movies['genres'] = movies['genres'].str.split('|')
movies['genres_str'] = movies['genres'].apply(lambda x: ' '.join(x))

count_vect = CountVectorizer()
genre_matrix = count_vect.fit_transform(movies['genres_str'])
cosine_sim = cosine_similarity(genre_matrix, genre_matrix)

def recommend_content(movie_title, top_n=5):
    """Повертає top_n схожих фільмів за жанрами"""
    if movie_title not in movies['title'].values:
        return f"Фільм '{movie_title}' не знайдено"
    idx = movies[movies['title'] == movie_title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]
    movie_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movie_indices].tolist()

# -------------------------------
# 4. Collaborative (спрощений) через user-item матрицю
# -------------------------------
user_movie_matrix = ratings.pivot(index='userId', columns='movieId', values='rating').fillna(0)

# Косинусна схожість між користувачами
user_sim = cosine_similarity(user_movie_matrix)
user_sim_df = pd.DataFrame(user_sim, index=user_movie_matrix.index, columns=user_movie_matrix.index)

# -------------------------------
# 5. Гібридна рекомендаційна функція
# -------------------------------
def recommend_hybrid(user_id, top_n=5):
    """Проста гібридна рекомендація: схожі користувачі + content-based"""
    if user_id not in user_movie_matrix.index:
        return f"Користувач {user_id} не знайдений"

    # 1. Collaborative: знаходимо 3 найбільш схожих користувачів
    sim_users = user_sim_df[user_id].sort_values(ascending=False).iloc[1:4].index
    movies_watched = ratings[ratings['userId'] == user_id]['movieId'].tolist()

    # 2. Беремо топ фільми від схожих користувачів, яких цей користувач ще не бачив
    candidate_movies = ratings[ratings['userId'].isin(sim_users)]
    candidate_movies = candidate_movies[~candidate_movies['movieId'].isin(movies_watched)]
    top_movies_cf = candidate_movies.groupby('movieId')['rating'].mean().sort_values(ascending=False).index[:top_n]

    # 3. Content-based: для кожного рекомендованого фільму додаємо 1 схожий за жанрами
    recommended_titles = []
    for mid in top_movies_cf:
        mid = int(mid)  # гарантуємо int
        title = movies[movies['movieId'] == mid]['title'].values[0]
        recommended_titles.append(title)
        similar = recommend_content(title, top_n=1)
        recommended_titles.extend(similar)
        if len(recommended_titles) >= top_n:
            break

    # Видаляємо дублікати
    recommended_titles = list(dict.fromkeys(recommended_titles))
    return recommended_titles[:top_n]

# -------------------------------
# 6. Приклад рекомендацій
# -------------------------------
user_id = 1
print(f"Гібридні рекомендації для користувача {user_id}:")
print(recommend_hybrid(user_id))

# -------------------------------
# 7. Проста оцінка RMSE для collaborative частини
# -------------------------------
# Створимо train/test
train, test = train_test_split(ratings, test_size=0.2, random_state=42)
train_matrix = train.pivot(index='userId', columns='movieId', values='rating').fillna(0)

# Косинусна схожість між користувачами на train
train_user_sim = cosine_similarity(train_matrix)
train_user_sim_df = pd.DataFrame(train_user_sim, index=train_matrix.index, columns=train_matrix.index)

# Прогнозування: для кожного user-movie беремо середній рейтинг від схожих користувачів
preds = []
for _, row in test.iterrows():
    u = row['userId']
    m = int(row['movieId'])
    if u in train_matrix.index:
        # схожі користувачі
        sim_users = train_user_sim_df.loc[u].sort_values(ascending=False).iloc[1:4].index
        ratings_sim = train_matrix.loc[sim_users, m].values if m in train_matrix.columns else []
        ratings_sim = [r for r in ratings_sim if r > 0]
        if len(ratings_sim) > 0:
            pred = np.mean(ratings_sim)
        else:
            pred = train_matrix.values.mean()
    else:
        pred = train_matrix.values.mean()
    preds.append(pred)

mse = mean_squared_error(test['rating'], preds)
rmse = np.sqrt(mse)

print(f"\nRMSE collaborative part: {rmse:.4f}")
```
